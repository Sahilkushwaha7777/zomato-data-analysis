```python
#question which solve to zomato dataset
#1)what type of resturant do the majority of customer order from?
#2)how many votes are each type of resturant recieved from customers?
#3)what are the rating majority customer recived?
#4)zomato has ovserved the most couples order most of thier food online. what is their average speding on each order?
#5)which mode(online or offline) has recieved the maximum rating?
#6)which type of resturant recieved more offline orders, so that zomato can pay customer with some good offers?
```

```python
#importing all required  library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create dataframe
dataframe = pd.read_csv("Zomato data .csv")
```

```python
dataframe
```

```python
#convert the data type of column - Rate

def handleRate(value):
  value = str(value).split("/")
  value= value[0];
  return float(value)

dataframe["rate"] = dataframe["rate"].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
#1)what type of resturant do the majority of customer order from?

dataframe.head()
```

```python
sns.countplot(x = dataframe["listed_in(type)"], hue = dataframe["listed_in(type)"], legend=False)
plt.xlabel("Type of resturant")
plt.ylabel("Count") # Added a y-axis label for clarity
plt.title("Count of Restaurant Types") # Added a title
```


#Conclusion - Majority of the resturant falls in dinning category

```python
#2)how many votes are each type of resturant recieved from customers?

dataframe.head()
```

```python
grouped_data = dataframe.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({"votes":grouped_data})
plt.plot(result, c = "green", marker= "o")
plt.xlabel("Types of resturant", c = "red",size = 20)
plt.ylabel("Votes", c= "r", size = 20)
```

#conclusion - dinning resturant has recievved maximum votes

```python
#questio3:- what are the rating majority customer recived?


plt.hist(dataframe['rate'],bins= 5)
plt.title("rating distribution")
plt.show()
```

#conclusion - the majority resturant recieved rating from 3.5 to 4

```python
dataframe.head()
```

```python
#question4 - zomato has ovserved the most couples order most of thier food online. what is their average spending on each order?\


couple_data = dataframe["approx_cost(for two people)"]
sns.countplot(x = couple_data, hue = dataframe["approx_cost(for two people)"], legend = False, palette="dark:red")
```

#conclusion - the majority of couples preffer resturant with an approximate cost of 300 rupees

```python
#question5) which mode(online or offline) has recieved the maximum rating?

plt.figure(figsize = (6,6))
sns.boxplot(x= 'online_order', y= 'rate', data = dataframe, hue = 'online_order')
plt.show()
```

#conclusion - offline order recieve lower rating on compare online order

```python
#question 6) - which type of resturant recieved more offline orders, so that zomato can pay customer with some good offers?

pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns ='online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu",fmt = 'd')
plt.title("heatmap")
plt.xlabel('Online ordr')
plt.ylabel("listed_in(type)")
plt.show()
```

#conclusion - dinning resturant primarily accept offline  orders , where as cafe primarily recieved online orders. this suggest that the client preffer orders in person at resturants, but prefer online ordering at cafes

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 15))

# Plot 1: Count of Restaurant Types
sns.countplot(x=dataframe["listed_in(type)"], ax=axes[0, 0], hue = dataframe["listed_in(type)"], legend = False)
axes[0, 0].set_xlabel("Type of restaurant")
axes[0, 0].set_ylabel("Count")
axes[0, 0].set_title("Count of Restaurant Types")

# Plot 2: Votes by Restaurant Type
axes[0, 1].plot(result, c="green", marker="o")
axes[0, 1].set_xlabel("Types of restaurant", c="red", size=20)
axes[0, 1].set_ylabel("Votes", c="r", size=20)
axes[0, 1].set_title("Votes by Restaurant Type")

# Plot 3: Rating Distribution
axes[1, 0].hist(dataframe['rate'], bins=5)
axes[1, 0].set_title("Rating Distribution")

# Plot 4: Approximate Cost Distribution for Couples
sns.countplot(x=couple_data, hue = dataframe["approx_cost(for two people)"], legend = False, palette="dark:red", ax=axes[1, 1])
axes[1, 1].set_title("Approximate Cost Distribution for Couples")
axes[1, 1].set_xlabel("approx_cost(for two people)")
axes[1, 1].set_ylabel("Count")

# Plot 5: Rating by Online Order
sns.boxplot(x= 'online_order', y= 'rate', data = dataframe, hue = 'online_order', ax=axes[2, 0], legend=False)
axes[2, 0].set_title("Rating by Online Order")
axes[2, 0].set_xlabel("Online Order")
axes[2, 0].set_ylabel("Rate")


# Plot 6: Offline Orders by Restaurant Type Heatmap
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu",fmt = 'd', ax=axes[2, 1])
axes[2, 1].set_title("Offline Orders by Restaurant Type")
axes[2, 1].set_xlabel('Online order')
axes[2, 1].set_ylabel("listed_in(type)")


plt.tight_layout()
plt.show()
```
